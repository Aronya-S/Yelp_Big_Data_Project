#### Checkin

```python
bucket = 'my-big-data-project-as/landing/'
file_path = 'gs://' + bucket + 'yelp_academic_dataset_checkin.json'
```

```python
sdf_checkin = spark.read.json(file_path)
```

```python
sdf_checkin.columns
```

```python
sdf_checkin.show(10)
```

```python
sdf_checkin.select([count(when(isnull(c), c)).alias(c) for c in ["business_id", "date"] ]).show()
```

```python
print(sdf_checkin.dtypes)
```

```python
gcs_bucket = 'my-big-data-project-as/'
output_file_path= f"gs://{gcs_bucket}/cleaned/cleaned_yelp_academic_dataset_checkin.parquet"
sdf_checkin.write.parquet(output_file_path)
```
