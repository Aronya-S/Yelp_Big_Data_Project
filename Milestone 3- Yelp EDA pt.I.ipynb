```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
pd.set_option('display.float_format', '{:.2f}'.format)
pd.set_option('display.width', 1000)
import seaborn as sns
```

```python
filepath = "gs://my-big-data-project-as/landing" 
```

#### Tip

```python
filename_tips = ['yelp_academic_dataset_tip.json']
```

```python
def perform_EDA(df : pd.DataFrame, filename : str):
    print(f"{filename} Number of records:")
    print(df.count())
    print(f"{filename} Number of duplicate records: { len(df)-len(df.drop_duplicates())}" )
    print(f"{filename} Info")
    print(df.info())
    print(f"{filename} Describe")
    print(df.describe())
    print(f"{filename} Columns with null values")
    print(df.columns[df.isnull().any()].tolist())
    rows_with_null_values = df.isnull().any(axis=1).sum()
    print(f"{filename} Number of Rows with null values: {rows_with_null_values}" )
    data_type = df.dtypes
    print(f"{filename} Data types: {data_type}" ) 
```

```python
for filename in filename_tips:
    print(f"Working on file: {filename}")
    tips_df = pd.read_json(f"{filepath}/{filename}", lines = True)
    perform_EDA(tips_df,filename)
```

```python
max_date = tips_df['date'].max()
print("Newest date:", max_date)
min_date = tips_df['date'].min()
print("Oldest date:", min_date)
print("\t")
print("Number of characters in each row:")
tips_df['Number_of_Characters'] = tips_df["text"].str.len()
tips_df['Number_of_Characters']
```

```python
tips_df['Number_of_Words'] = tips_df["text"].str.split().str.len()
tips_df['Number_of_Words']
```

```python
tips_df.tail()
```

```python
num_rows_with_text = len(tips_df[tips_df['text'].str.len() > 0])

print("Number of rows with text:", num_rows_with_text)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(tips_df['Number_of_Characters'], tips_df['Number_of_Words'], color='purple')
plt.xlabel('Number of Characters')
plt.ylabel('Number of Words')
plt.title('Comparison of Amount of Words vs. Characters Written in Text')
plt.show()
```

#### Business

```python
def perform_EDA2(df : pd.DataFrame, filename : str):
    print(f"{filename} Number of records:")
    print(df.count())
    print(f"{filename} Info")
    print(df.info())
    print(f"{filename} Describe")
    print(df.describe())
    print(f"{filename} Columns with null values")
    print(df.columns[df.isnull().any()].tolist())
    rows_with_null_values = df.isnull().any(axis=1).sum()
    print(f"{filename} Number of Rows with null values: {rows_with_null_values}" )
    data_type = df.dtypes
    print(f"{filename} Data types: {data_type}" ) 
```

```python
filename_business = ['yelp_academic_dataset_business.json']
for filename in filename_business:
    print(f"Working on file: {filename}")
    business_df = pd.read_json(f"{filepath}/{filename}", lines = True)
    perform_EDA2(business_df,filename)
```

```python
dp = sns.displot(business_df.stars)
```

```python
plt.hist(business_df['stars'], bins = [1,1.5,2,2.5,3,3.5,4,4.5,5])
plt.xlabel("Stars")
plt.ylabel("Count")
plt.show()
```

```python
cities = business_df['city'].value_counts().reset_index()
cities.columns = ['city', 'count']
ordered = cities.sort_values(by='count', ascending=False)
top_10_cities = ordered.head(10)
plt.figure(figsize=(12, 8))

sns.barplot(x='city', y='count', data=top_10_cities)
plt.title('Top 10 Cities Found From Dataset')
plt.xlabel('City')
plt.ylabel('Number of Businesses')
plt.xticks()
plt.tight_layout()
plt.show()
```
